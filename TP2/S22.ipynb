```python
!pip install pyspark
!apt install openjdk-8-jdk-headless -qq
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"

from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import SparkSession

from google.colab import drive
drive.mount('/content/drive')

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
sqlContext = SQLContext(sc)

movies = sqlContext.read.parquet('/content/drive/MyDrive/Colab Notebooks/TP2/parquet/movies.parquet').rdd
companies = sqlContext.read.parquet('/content/drive/MyDrive/Colab Notebooks/TP2/parquet/companies.parquet').rdd
```

```python
max_popularity = movies.filter(lambda x: x.popularity != None).map(lambda x: x.popularity).max()
movie_max_popularity = movies.filter(lambda x: x.popularity == max_popularity).first()

name = movie_max_popularity.title
companies_id = movie_max_popularity.production_companies.split(',')
company1_id = companies_id[0]
company2_id = companies_id[1]

company1_name = companies.filter(lambda x: str(x.id) == company1_id).first().name
company2_name = companies.filter(lambda x: str(x.id) == company2_id).first().name

print("Película", name, "con popularidad", max_popularity)
print("Producida por la compañía", company1_name, "de ID", company1_id)
print("Producida por la compañía", company2_name, "de ID", company2_id)
```
